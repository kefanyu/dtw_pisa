```python
sci = pd.read_excel("../../data/avg/new_trend.xlsx",sheet_name='Science')
sci.drop(['Unnamed: 0','se_2000','se_2003','se_2006','se_2009','se_2012','se_2015','se_2018','rank_2000','rank_2003','rank_2006','rank_2009','rank_2012','rank_2015','rank_2018'], axis=1, inplace=True)
sci.rename({'avg_2000': "2000",'avg_2003': "2003", 'avg_2006': "2006", 'avg_2009': "2009",'avg_2012': "2012",'avg_2015': "2015",'avg_2018': "2018"}, axis=1, inplace=True)
#reading = reading.drop(index=[259])

sci
```

```python
reading = reading.iloc[1:]
reading
```

```python
reading_val = reading.drop(["Countries"],axis=1)
reading_lst = []

# Iterate through the rows of the DataFrame
for index, row in reading_val.iterrows():
    row_list = row.tolist()
    reading_lst.append(row_list) 

# Now, big_list contains the lists corresponding to the rows of the DataFrame
print(reading_lst)
```

```python
reading_mean = []

# Calculate the mean of each column and append to the list
for column_name in reading_val.columns:
    column_mean = reading_val[column_name].mean()
    reading_mean.append(column_mean)

print(reading_mean)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

x = ['2000', '2003', '2006', '2009', '2012', '2015','2018']

# Loop through each list and create a line plot
for sublist in reading_lst:
    #x = list(range(1, len(sublist) + 1))  # X-axis values
    sns.lineplot(x=x, y=sublist, color = "pink")

sns.lineplot(x=x, y=reading_mean, color = "red",linewidth=2.5,label = "All Countries")

plt.xlabel('Year')
plt.ylabel('Average')
plt.legend(title='Average of Reading')
plt.savefig('../../plots/all_country/all_country_avg_reading.png')
plt.show()
```
