```python
import pandas as pd
import numpy as np
import xlrd
```

```python
reading = pd.read_excel("../../data/oecd/oecd_2022_clean.xlsx",sheet_name='Reading')
math = pd.read_excel("../../data/oecd/oecd_2022_clean.xlsx",sheet_name='Math')
sci = pd.read_excel("../../data/oecd/oecd_2022_clean.xlsx",sheet_name='Science')

reading
```

```python
reading.drop(['Unnamed: 0'], axis=1, inplace=True)
math.drop(['Unnamed: 0'], axis=1, inplace=True)
sci.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
reading_dtw = reading.drop(['Countries'], axis=1)
math_dtw = math.drop(['Countries'], axis=1)
sci_dtw = sci.drop(['Countries'], axis=1)
```

```python
reading_dtw = reading_dtw.ffill(axis=1).bfill(axis=1)
math_dtw = math_dtw.ffill(axis=1).bfill(axis=1)
sci_dtw = sci_dtw.ffill(axis=1).bfill(axis=1)
```

```python
reading_dtw.insert(0, 'Country', reading['Countries'].tolist())
math_dtw.insert(0, 'Country', reading['Countries'].tolist())
sci_dtw.insert(0, 'Country', reading['Countries'].tolist())
reading_dtw
```

```python
avg_reading = np.mean(reading_dtw, axis=0).tolist()
avg_math = np.mean(math_dtw, axis=0).tolist()
avg_sci = np.mean(sci_dtw, axis=0).tolist()
```

```python
std_reading = np.std(reading_dtw, axis=0).tolist()
std_math = np.std(math_dtw, axis=0).tolist()
std_sci = np.std(sci_dtw, axis=0).tolist()
```

```python
normal_reading = reading_dtw.copy()
normal_math = math_dtw.copy()
normal_sci = sci_dtw.copy()
```

```python
normal_reading['2000'] = (normal_reading['2000'] - avg_reading[0]) / std_reading[0]
normal_reading['2003'] = (normal_reading['2003'] - avg_reading[1]) / std_reading[1]
normal_reading['2006'] = (normal_reading['2006'] - avg_reading[2]) / std_reading[2]
normal_reading['2009'] = (normal_reading['2009'] - avg_reading[3]) / std_reading[3]
normal_reading['2012'] = (normal_reading['2012'] - avg_reading[4]) / std_reading[4]
normal_reading['2015'] = (normal_reading['2015'] - avg_reading[5]) / std_reading[5]
normal_reading['2018'] = (normal_reading['2018'] - avg_reading[6]) / std_reading[6]
normal_reading['2022'] = (normal_reading['2022'] - avg_reading[7]) / std_reading[7]
```

```python
normal_reading
```

```python
normal_math['2000'] = (normal_math['2000'] - avg_math[0]) / std_math[0]
normal_math['2003'] = (normal_math['2003'] - avg_math[1]) / std_math[1]
normal_math['2006'] = (normal_math['2006'] - avg_math[2]) / std_math[2]
normal_math['2009'] = (normal_math['2009'] - avg_math[3]) / std_math[3]
normal_math['2012'] = (normal_math['2012'] - avg_math[4]) / std_math[4]
normal_math['2015'] = (normal_math['2015'] - avg_math[5]) / std_math[5]
normal_math['2018'] = (normal_math['2018'] - avg_math[6]) / std_math[6]
normal_math['2022'] = (normal_math['2022'] - avg_math[7]) / std_math[7]
```

```python
normal_sci['2000'] = (normal_sci['2000'] - avg_sci[0]) / std_sci[0]
normal_sci['2003'] = (normal_sci['2003'] - avg_sci[1]) / std_sci[1]
normal_sci['2006'] = (normal_sci['2006'] - avg_sci[2]) / std_sci[2]
normal_sci['2009'] = (normal_sci['2009'] - avg_sci[3]) / std_sci[3]
normal_sci['2012'] = (normal_sci['2012'] - avg_sci[4]) / std_sci[4]
normal_sci['2015'] = (normal_sci['2015'] - avg_sci[5]) / std_sci[5]
normal_sci['2018'] = (normal_sci['2018'] - avg_sci[6]) / std_sci[6]
normal_sci['2022'] = (normal_sci['2022'] - avg_sci[7]) / std_sci[7]
```

```python
import openpyxl
```

```python
normal_reading.to_excel('../../data/oecd/oecd_2022_normal.xlsx', sheet_name='Reading')
```

```python
from openpyxl import load_workbook

ExcelWorkbook = load_workbook('../../data/oecd/oecd_2022_normal.xlsx')

writer = pd.ExcelWriter('../../data/oecd/oecd_2022_normal.xlsx', engine = 'openpyxl')

writer.book = ExcelWorkbook

normal_math.to_excel(writer, sheet_name = 'Math')
normal_sci.to_excel(writer, sheet_name = 'Science')
writer.save()
writer.close()
```
