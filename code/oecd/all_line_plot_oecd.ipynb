```python
import pandas as pd
import numpy as np
import xlrd
```

```python
reading = pd.read_excel("../../data/oecd/oecd_avg_clean.xlsx",sheet_name='Reading')
reading.drop('Unnamed: 0', axis=1, inplace=True)
reading
```

```python
math = pd.read_excel("../../data/oecd/oecd_avg_clean.xlsx",sheet_name='Math')
math.drop('Unnamed: 0', axis=1, inplace=True)
math
```

```python
sci = pd.read_excel("../../data/oecd/oecd_avg_clean.xlsx",sheet_name='Science')
sci.drop('Unnamed: 0', axis=1, inplace=True)
sci
```

```python
reading_val = reading.drop(["Countries"],axis=1)
reading_lst = []

# Iterate through the rows of the DataFrame
for index, row in reading_val.iterrows():
    row_list = row.tolist()
    reading_lst.append(row_list) 

# Now, big_list contains the lists corresponding to the rows of the DataFrame
print(reading_lst)
```

```python
reading_mean = []

# Calculate the mean of each column and append to the list
for column_name in reading_val.columns:
    column_mean = reading_val[column_name].mean()
    reading_mean.append(column_mean)

print(reading_mean)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

x = ['2000', '2003', '2006', '2009', '2012', '2015','2018']

# Loop through each list and create a line plot
for sublist in reading_lst:
    #x = list(range(1, len(sublist) + 1))  # X-axis values
    sns.lineplot(x=x, y=sublist, color = "pink")

sns.lineplot(x=x, y=reading_mean, color = "red",linewidth=2.5,label = "OECD Countries")

plt.xlabel('Year')
plt.ylabel('Average')
plt.legend(title='Average of Reading')
plt.savefig('../../plots/all_country/oecd_avg_reading.png')
plt.show()
```

```python
math_val = math.drop(["Countries"],axis=1)
math_lst = []
math_mean = []

# Iterate through the rows of the DataFrame
for index, row in math_val.iterrows():
    row_list = row.tolist()
    math_lst.append(row_list)

# Calculate the mean of each column and append to the list
for column_name in math_val.columns:
    column_mean = math_val[column_name].mean()
    math_mean.append(column_mean)
```
