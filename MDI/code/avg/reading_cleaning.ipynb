```python
readings_avg2 = pd.melt(readings_avg, id_vars=['Countries'], value_vars=['2000','2003','2006', '2009','2012','2015','2018'],   var_name='Year', value_name='avg', ignore_index=False)
readings_avg2
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
con2 = readings_avg2['Countries'].unique().tolist()
cat2  = readings_avg2['Year'].unique().tolist()

fig = go.Figure()

data2 = []

for i in con2:

    fig.add_trace(  # Add A trace to the figure
            go.Scatter(  # Specify the type of the trace
                x=readings_avg2.loc[readings_avg2['Countries'] == i]['Year'].tolist(),  # Data-x
                y=readings_avg2.loc[readings_avg2['Countries'] == i]['avg'].tolist(),  # Data-y
                mode="lines+markers",
                # note the re-normalization of population to map to width to units of "pixels"
                marker=dict(
                    color=readings_avg2.loc[readings_avg2['Countries'] == i]['avg'].tolist(),
                    showscale=True,
                    colorscale="Viridis",
                    symbol="circle",
                ),
                opacity=1.0,
                visible=False,  # Specify whether or not to make data-visible when rendered
            )
        )
    


#visual = [False]*len(con)
bton2 = []

for j in range(0, len(con2)):
    visual2 = [False]*len(con2)
    visual2[j] = True
    #print(visual)

    bton2.append(
        dict(label=con2[j],
                     method="update",
                     args=[{"visible": visual2},
                           {"title": f"reading test score average for {con2[j]}"}])
    )

fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=list(bton2),
        )
    ])
  
fig.show()
fig.write_html("reading_trend.html")
```
